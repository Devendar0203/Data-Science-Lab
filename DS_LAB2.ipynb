```python
import pandas as pd
import numpy as np
# sample dataset
df=pd.DataFrame({'Age':[25,30,np.nan,40,35],
  'Department':['HR','Finance','Finanace',np.nan,'It']})
df['Age'].fillna(df['Age'].mean(), inplace=True)
#mode for categorical
df['Department'].fillna(df['Department'].mode()[0],inplace=True)
print(df)
```

#forward fill


```python
#forward fill
import pandas as pd
import numpy as np
# Sample dataset
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df_ffill = df.copy()
df_ffill.fillna(method='ffill', inplace=True)
print(df_ffill)

```

#backward fill

```python
#backward fill
import pandas as pd
import numpy as np
# Sample dataset
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df_bfill = df.copy()
df_bfill.fillna(method='bfill', inplace=True)
print(df_bfill)

```

#Deletion

```python
#Deletion
# Drop rows with missing values
import pandas as pd
import numpy as np
# Sample dataset
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df_drop_rows = df.dropna()
print("After dropping rows:\n", df_drop_rows)

# Drop columns with missing values
import pandas as pd
import numpy as np
# Sample dataset
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df_drop_cols = df.dropna(axis=1)
print("After dropping columns:\n", df_drop_cols)
```

Removing Duplicates

1️⃣ Exact Match Removal

```python
import pandas as pd

# Sample dataset with duplicates
df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

print("Original Data:\n", df)

# Remove exact duplicates
df_exact = df.drop_duplicates()
print("\nAfter Exact Match Removal:\n", df_exact)
```

#Subset-Based Removal

```python
#Subset-Based Removal
# Remove duplicates based only on 'ID'
df_subset_id = df.drop_duplicates(subset=['ID'])
print("\nAfter Subset-Based Removal (ID):\n", df_subset_id)
# Remove duplicates based only on 'Name'
df_subset_name = df.drop_duplicates(subset=['Name'])
print("\nAfter Subset-Based Removal (Name):\n", df_subset_name)
```

#Date Standardization

```python
#Correcting Inconsistent Formats
#Date Standardization

# Sample dataset with inconsistent date formats
df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

# Convert all dates to standard ISO format (YYYY-MM-DD)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

print(df)

```

#Case Normalization

```python
#Case Normalization
# Sample dataset with inconsistent text cases
df = pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

# Convert all names to lowercase
df['Name_lower'] = df['Name'].str.lower()

# Convert all names to uppercase
df['Name_upper'] = df['Name'].str.upper()

print(df)
```
