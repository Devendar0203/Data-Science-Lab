```python
import numpy as np
from scipy.spatial import distance
pointA=np.array([2,4,6])
pointB=np.array([5,1,9])
#Euclidean distance
euclidean_dist = distance.euclidean(pointA,pointB)
print("Euclidean Distance:",euclidean_dist)
#similarity(inverse of distance)
similarity_euclidean=1/(1+euclidean_dist)
print("Similarity(inverse of distance):",similarity_euclidean)
```

2️⃣ Manhattan (City Block) Distance

```python
import numpy as np
from scipy.spatial import distance
manhattan_dist=distance.cityblock(pointA,pointB)
print("Manhattan Distance:",manhattan_dist)
#similarity (inverse of distance)
similarity_manhattan = 1 / (1 + manhattan_dist)
print("Manhattan Similarity:", similarity_manhattan)
```


3️⃣ Minkowski Distance (generalized form)
Manhattan Distance measures distance by adding absolute differences.
 #it is general form it can become manhattan when p=1 it can become euclidean when p=2

```python
#it is general form it can become manhattan when p=1 it can become euclidean when p=2
import numpy as np
from scipy.spatial import distance
minkowski_dist_p3 = distance.minkowski(pointA, pointB, p=3)
print("Minkowski Distance (p=3):", minkowski_dist_p3)

# Similarity (inverse of distance)
similarity_minkowski = 1 / (1 + minkowski_dist_p3)
print("Minkowski Similarity (p=3):", similarity_minkowski)
```

2.Statistical Measures

```python
#Pearson Correlation: Measures linear relationship between two variables (range: -1 to +1).
import pandas as pd
from scipy.stats import pearsonr

# Load the dataset
tips = pd.read_csv("tips.csv")

# Example: Pearson correlation between 'total_bill' and 'tip'
corr, p_value = pearsonr(tips['total_bill'], tips['tip'])

print(f"Pearson correlation coefficient: {corr}")
print(f"P-value: {p_value}")
```

Spearman Rank Correlation: Non-parametric measure based on rank order.

```python
import pandas as pd
from scipy.stats import spearmanr
#example dataset
df=pd.DataFrame({
    'X':[10,20,30,40,50],
    'Y':[12,18,33,47,55]
})
#Spearman correlation coefficent and p-value
corr_value, p_value=spearmanr(df['X'],df['Y'])
print(f"Spearman correlation coefficient:{corr_value}")
print(f"P-value: {p_value}")
#for dataset:
#df=pd.read_csv("tips.csv")
#print(df.corr(method='spearman'))
```

3. Edit-Based Measures (String/Text Data)

#Hamming **Distance**

```python
#Hamming Distance
def hamming_distance(str1,str2):
    if len(str1) != len(str2) :
        raise ValueError("Strings must be of equal length")
    #count differing positions
    return sum(ch1 != ch2 for ch1, ch2 in zip(str1, str2))
#example usage
s1 = "car"
s2 = "cat"

dist=hamming_distance(s1,s2)
print(f"Hamming distance between '{s1}' and '{s2}' : {dist}")
```

**4. Token-Based Measures (Sets/Text)**

```python
#jaccard index
def jaccard_index(str1,str2):
  set1,set2=set(str1.split()),set(str2.split())
  intersection = set1.intersection(set2)
  union=set1.union(set2)
  return len(intersection) / len(union)

s1="data science is fun"
s2="science makes data usefull"
print("Jaccard Index:",jaccard_index(s1,s2))
```

**2️⃣ Cosine Similarity**

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

s1 ="data science fun"
s2="science makes data useful"
#convert text to vector reperesentation
vectorizer = CountVectorizer().fit([s1,s2])
vectors=vectorizer.transform([s1,s2])
#compute cosine similarity
cos_sim = cosine_similarity(vectors[0],vectors[1])[0][0]
print("Cosine similarity:",cos_sim)
```

```python
def lcs_length(X, Y):
    m, n = len(X), len(Y)

    # Create matrix
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Build matrix
    for i in range(m):
        for j in range(n):
            if X[i] == Y[j]:
                dp[i + 1][j + 1] = dp[i][j] + 1
            else:
                dp[i + 1][j + 1] = max(dp[i][j + 1], dp[i + 1][j])

    return dp[m][n]


# Example usage
seq1 = "ABCDEF"
seq2 = "AEBDF"

length = lcs_length(seq1, seq2)

print("Longest Common Subsequence Length:", length)
```
